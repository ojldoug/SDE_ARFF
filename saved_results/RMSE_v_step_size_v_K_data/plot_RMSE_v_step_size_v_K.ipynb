```python
import numpy as np
import matplotlib.pyplot as plt

ex_name = "ex3-1D_cubic"
K = [16, 32, 64, 128, 256]

# Set up both figures and axes
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

for k in K:
    data = np.load(f'{ex_name}_K{k}.npz')
    step_size = data["step_size"]
    drift_RMSE = data["drift_RMSE"]
    diffusion_RMSE = data["diffusion_RMSE"]

    ax1.loglog(step_size, drift_RMSE, label=f'K={k}')
    ax2.loglog(step_size, diffusion_RMSE, label=f'K={k}')

# Label and show drift RMSE plot
ax1.set_xlabel('step size')
ax1.set_ylabel('drift RMSE')
ax1.legend()
fig1.show()

# Label and show diffusion RMSE plot
ax2.set_xlabel('step size')
ax2.set_ylabel('diffusion RMSE')
ax2.legend()
fig2.show()
```
