```python
import numpy as np
import matplotlib.pyplot as plt

ex_name = "ex3-1D_cubic"
K = [128]

for k in K:
    # load
    data = np.load(f'{ex_name}_K{k}.npz')
    step_size = data["step_size"]
    drift_RMSE = data["drift_RMSE"]
    diffusion_RMSE = data["diffusion_RMSE"]


    # plot
    plt.loglog(step_size, drift_RMSE)
    plt.xlabel('step size')
    plt.ylabel('drift RMSE')
    plt.show()
    
    plt.loglog(step_size, diffusion_RMSE)
    plt.xlabel('step size')
    plt.ylabel('diffusion RMSE')
    plt.show()
```
