```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

script_dir = os.getcwd()
data_dir = os.path.join(script_dir, 'loss_v_time_data')

#test_names = ['1 - 1d sde', '2 - 2d sde', '3 - 1d sde-cubic', '3 - 2d sde-cubic', '4 - 3d sde-LT', '4 - 3d sde-spd', '6 - SIRGillespie', '8 - nonGaussian']
test_names = ['3-1d_sde-cubic']
n_subsample = 1000
save = False

for i in range(len(test_names)):
    test_name = test_names[i]

    # load Adam
    Adam_path = os.path.join(data_dir, f'example{test_name}_SS{n_subsample}.csv')
    data = pd.read_csv(Adam_path)
    
    Adam_time = data.iloc[:, 0]
    Adam_loss = data.iloc[:, 1]
    Adam_time_above_std = data.iloc[:, 2]
    Adam_time_below_std = data.iloc[:, 3]
    Adam_loss_above_std = data.iloc[:, 4]
    Adam_loss_below_std = data.iloc[:, 5]
    
    # load ARFF
    ARFF_path = os.path.join(data_dir, f'ex{test_name}_SS{n_subsample}.txt')
    with open(ARFF_path, 'r') as file:
        line = file.readline().strip()
        ARFF_time, ARFF_loss, ARFF_time_above_std, ARFF_time_below_std, ARFF_loss_above_std, ARFF_loss_below_std = map(float, line.split(','))

    # load min loss
    min_loss_path = os.path.join(data_dir, f'ex{test_name}_SS{n_subsample}_min_loss.txt')
    with open(min_loss_path, 'r') as file:
        line = file.readline().strip()
        min_loss = float(line)

    # create plot
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # plot min loss
    ax.axhline(y=min_loss, color='green', linestyle='--', linewidth=1.5, label='Expected min loss')

    # plot Adam
    ax.errorbar(
        Adam_time, Adam_loss,
        xerr=[Adam_time_below_std, Adam_time_above_std], 
        yerr=[Adam_loss_below_std, Adam_loss_above_std],  
        fmt='o', color='blue', ecolor='blue', elinewidth=1.0, capsize=3, label='Adam', zorder=1
    )

    # plot ARFF
    ax.errorbar(
        ARFF_time, ARFF_loss,
        xerr=[[ARFF_time_below_std], [ARFF_time_above_std]], 
        yerr=[[ARFF_loss_below_std], [ARFF_loss_above_std]],  
        fmt='o', color='red', ecolor='red', elinewidth=1.0, capsize=3, label='ARFF', zorder=4
    )

        # load Adam deep
    Adam_path = os.path.join(data_dir, f'example{test_name}_deep_SS{n_subsample}.csv')
    data = pd.read_csv(Adam_path)
    
    Adam_time = data.iloc[:, 0]
    Adam_loss = data.iloc[:, 1]
    Adam_time_above_std = data.iloc[:, 2]
    Adam_time_below_std = data.iloc[:, 3]
    Adam_loss_above_std = data.iloc[:, 4]
    Adam_loss_below_std = data.iloc[:, 5]
    
    # load ARFF no resampling
    ARFF_path = os.path.join(data_dir, f'ex{test_name}_noResampling_SS{n_subsample}.txt')
    with open(ARFF_path, 'r') as file:
        line = file.readline().strip()
        ARFF_time, ARFF_loss, ARFF_time_above_std, ARFF_time_below_std, ARFF_loss_above_std, ARFF_loss_below_std = map(float, line.split(','))

    # plot Adam deep
    ax.errorbar(
        Adam_time, Adam_loss,
        xerr=[Adam_time_below_std, Adam_time_above_std], 
        yerr=[Adam_loss_below_std, Adam_loss_above_std],  
        fmt='o', color='skyblue', ecolor='skyblue', elinewidth=1.0, capsize=3, label='Adam deep', zorder=2
    )

    # plot ARFF no resampling
    ax.errorbar(
        ARFF_time, ARFF_loss,
        xerr=[[ARFF_time_below_std], [ARFF_time_above_std]], 
        yerr=[[ARFF_loss_below_std], [ARFF_loss_above_std]],  
        fmt='o', color='orange', ecolor='orange', elinewidth=1.0, capsize=3, label='ARFF no resampling', zorder=3
    )

    
    # labels
    ax.set_xlabel('Time (s)', fontsize=14)
    ax.set_ylabel('Validation Loss', fontsize=14)
    ax.legend(fontsize=12)
    ax.grid(True)
    
    plt.show()
    
    if save:
        output_dir = os.path.join(script_dir, 'loss_v_time_plots')
        output_path = os.path.join(output_dir, f'ex{test_name}_SS{n_subsample}.png')
        fig.savefig(output_path, dpi=300, bbox_inches='tight')
        plt.close(fig)
        
    

    
```
